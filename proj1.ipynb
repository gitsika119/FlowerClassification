```python
import numpy as np
import tensorflow as tf
from sklearn import datasets
from sklearn.svm import SVC
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
import pandas as pd

iris=datasets.load_iris()
X=iris.data
Y=iris.target
iris_frame=pd.DataFrame(iris.data,columns=iris.feature_names)
iris_frame.head()

x_axis=1
y_axis=3
plt.scatter(X[:,x_axis],X[:,y_axis],c=Y)
plt.xlabel(iris.feature_names[x_axis])
plt.ylabel(iris.feature_names[y_axis])
plt.show()

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
# Initialize the model (using a SVM as an example)
model = SVC()
print()
# Train the model on the training data
model.fit(x_train, y_train)
# Test the model on the testing data and print accuracy
accuracy = model.score(x_test, y_test)
print(f"Model accuracy/score: {accuracy * 100:.2f}%\n")
# Make a prediction on the test set
y_pred = model.predict(x_test)
print("Predicted classes:", y_pred)
```
